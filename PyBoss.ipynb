```python
import os
import csv
from pathlib import Path  
filepath = Path("employee_data.csv")
with open(filepath, newline="", encoding='utf-8' ) as csvfile:
    csvreader = csv.reader(csvfile, delimiter=",")
    reader = csv.reader(csvfile)
    next(reader, None)
    emp_id = []
    name = []
    DOB = []
    SSN = []
    state = []
    for row in csvreader:
        emp_id.append(row[0])
        name.append(row[1])
        DOB.append(row[2])
        SSN.append(row[3])
        state.append(row[4])
#print(name)
```
